# Exploring Spotify Songs from 1960-2020

## Introduction

Music plays a central role in global culture. The past century of music is punctuated by constant shifts due to the proliferation of unique musical genres and the introduction of new technologies that fundamentally forced the music industry to adapt. From the Turntable to the iPod, the affordances of each technology have influenced the kinds of music artists are able to create and the scale of audience they are able to reach. Yet, with increased access to music, competition among artists has drastically increased along with changing business models. The desire and need to create a "hit" song has become critical to one artist's success over another, and research shows that our attention spans have significantly dropped since the emergence of new digital technologies. Our research is looking to explore how key qualities of songs have changed over the past 80 years, especially with the introduction of new music technologies that have fundamentally altered the model of the music industry. We specifically are interested in considering if the duration of songs has meaningfully decreased over the past 80 years as song length becomes less of a determinator of song success and revenue with the rise of streaming, which pays on an individual 'stream' basis. Do artists not care about making full length albums or long songs anymore if they can make a hit song like 'Old Town Road' which lasts 1 minute and 53 seconds. We also would like to analyze how new genres and decades have changed the danceability, tempo, and popularity of music, which also can play a role in the changing qualities of songs over time. 

<b> Main Findings: </b>

- Between 1960 and 2020, average song duration decreased. We can support this finding by considering the slope coefficient of -6.84 between 1960 and 2020. Additionally, from the late 1960s to the early 1970s there was an increase in song duration, which most notably happens to be the year that the 8-track tape and the Turntable were released. Both of these technologies allowed people to play songs all the way through, or listen to records in their entirety - and sales in the music industry were dependent more on entire record sales, rather than specific songs. In other words, artists would be more focused on making the best records to sell rather than indvidual songs, so song length wasn't as large of a factor in sales. A similar but opposite trend showing the largest decrease in song duration is shown from 2000 to the late 2000's, which is when the invention of new streaming services like Napstr, Itunes, and Myspace were brought to market as well as the invention of new music playing inventions like the IPod. These technologies lend themselves to the streaming model which pays artists money for the individual stream of a specific song, regardless of length. 

- There is not a strong linear correlation, however, between the year a song is released and the duration of the song. Even though the overall average song duration decreased over 80 years, when we performed a linear regression on song duration and year, we calculated an r squared of 0.000002 or approximately 0. This finding communicates to us that the independent variable (year) cannot be predicted by the depedent variable (duration). To further support this claim when we performed a multi-variable analysis using danceability, duration, energy, and tempo to predict the year a song was released, we received a prediction accuracy rating of 3.4%. However, when we attempted to predict the year a song was released within 5 years (plus or minus) we were 23.1% accurate. Our accuracy scores demonstrates that the qualities of songs vary somewhat significantly from year to year, enough to make a prediction that is 12 times more accurate than a guess but still not entirely accurate. 

- Danceability has seen peaks in the late 1960s, the 1980s, and the late 1990s. To elaborate, danceability describes how 'suitable' a song is for dancing and is determined by the tempo, rhythm, and beat strength. The danceability is determined on a scale of 0.0 to 1.0 and is one of Spotify's audio metrics that they use to suggest/organize music. This is interesting because each one of these peaks corresponds to a notably popular genre of music: 1960s - Rock n Roll, 1980s - dance music, 1990s/early 2000s - Rock + Rap. Additionally, instrumentalness, or the amount of live instruments that are used in a recording, has decreased since the 1960s which can support the rise in dance/electronic music and therefore support the increase in danceability.

- Songs are getting sadder over time. Valence, which is a quality that measures how positive a song is has a strong negative coorelation of -0.75 with both popularity and year, which communicates botht that songs are getting sadder over time AND that sad songs are more popular that happy songs.


**Summary of results:**

After analyzing a dataset of over 120,000 Spotify songs, we can determine over the past 40 years songs are getting sadder, shorter, and less instrumental. Yet, music is more popular ever getting more streams every year due to population increase and the ubiquity of music sharing/streaming technology. From our analysis, song duration showed trends over time that were indicative of the ways that new technology has effected the music indsutry. For example, song duration was longest when Vinyl was released but experienced a sharp decline following the introduction of streaming and the IPod. Both these innovations directly impacted song duration the most compared to other qualities of songs. In addition, as song duration has decreased, danceability of songs has marginally increased over the past 40 years. However, there is not a strong linear relationship between year, duration, and danceability thus far in our analysis - demonstrated by r squared scores close to 0 in our linear regression models of 'years vs duration' and 'year vs danceability'.

**Preregistration Statement 1**

How does song duration relate to year?

Below we ran a regression on the duration of songs against release year. As stated above, we hoped to see if there was a linear relationship between song duration and year. Specifically, we expected song duration to decrease as year increased. 

```python
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

x = new_data[['year']]
y = new_data['duration_sec']

predict = LinearRegression().fit(x, y)

slope = predict.coef_[0]
intercept = predict.intercept_

print('The slope is:', slope)
print('The intercept is:', intercept)

sns.regplot(x, y, data=new_data).set_title('Duration vs. Year')

rsquared = predict.score(x, y)
print('the r squared value is:', rsquared)
```

**Evaluation of Significance for Preregistration Statement 1**

Our slope of -6.84 shows that between 1960 and 2020, song duration decreased. This downward trend supports our inital hypothesis that song duration has decreased over time as artists focus more on producing "hits" instead of longer, more flushed out pieces of work. However, our incredibly small r squared value close to 0 shows that there is not a strong linear relationship between song duration and year. This leads us to believe that a linear regression may not be the best model to compare song duration against year.

**Preregistration Statement 2**

Can we utilize qualities of songs to predict the year they were released?

Below we analyzed our second preregistration statement. In this analysis we were looking to see if we could use danceability, duration, energy, and tempo to predict the year a song was released. 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#Using a multivariable linear regression to see how danceability, duration, energy, and tempo are able to predict year
xfit = new_data[['danceability', 'duration_sec', 'energy', 'tempo']]
yfit = new_data['year']

multiregression = LinearRegression().fit(xfit, yfit)
print('The coefficients of the input variables are {}'.format(multiregression.coef_))
r_value = multiregression.score(xfit,yfit)
print('The r squared value is:', r_value)

#Using a logistic regression to see if we can successfully use these inputs to predict the year a song was made
spot_train, spot_test = train_test_split(new_data, test_size=.9, random_state=15)
multiregression = LogisticRegression(max_iter= 1000).fit(spot_train[['danceability', 'duration_sec', 'energy', 'tempo']], spot_train['year'])

correct = sum(spot_test['year'] == multiregression.predict(spot_test[['danceability', 'duration_sec', 'energy', 'tempo']]))
total = len(spot_test)
print("The prediction accuracy using the test set is: {:.1f}%".format(correct/total*100))
```

**Evaluation of Significance for Preregistration 2**

Our prediction accuracy of 3.4% shows that it was incredibly difficult for us to predict the release year of a song based on danceability, duration, energy, and tempo. 3.4% is around the same accuracy as if we were guessing, considering we would be guessing 1 year out of 56. (1/56 ~ 2%) This result may be explained by the accuracy of our regression. The r squared value of .107 is quite low and demonstrates that only 10.7% of our data in the regression fits the model.

We are not completely surprised by this result because the difference between a song released in 1988 and 1989 may be very minimal, however, if our model does not predict the year exactly, our model marks that as a failure. This may not be the most effective prediction method. Since songs produced within the same decade are generally similar and we want a model that is able to reflect that with its predictions, in the next cell, we try to use our model to predict songs within a decade (+ or - 5 years of the correct year).

```python
year_predictions = multiregression.predict(spot_test[['danceability', 'duration_sec', 'energy', 'tempo']])
spot_test['predicted_values'] = year_predictions
spot_test['year_plus'] = spot_test['year'] + 5
spot_test['year_minus'] = spot_test['year'] - 5
total_years = len(spot_test.query('predicted_values < year_plus and predicted_values > year_minus'))
correct_years = total_years / len(spot_test)
print("The prediction accuracy within 5 years of the correct year using the test set is: {:.1f}%".format(correct_years*100));
```

As expected, our model was much more accurate at predicting songs when given a range of within 5 years of their release. Although 23.1% is not necessarily pin-point accuracy, it still demonstrates that when accounting for the trends of changes in song qualities, we can make a better prediction than if we were simply guessing. From this analysis we can conclude that songs have a significant change in qualities over time that although isn't entirely predictable, is still significant.

When looking at these results it makes sense that year isn't correlated to duration of songs as much as we thought. Initially, we were trying to use year as a placeholder variable that existed within our dataset to demonstrate the rise of new technologies over time. For example, by using years from 1980s-2020 we can account for changes in song duration over a 40 year span where new technologies were obviously introduced. However, through our results it's clear that year as a standalone variable does not provide enough context to the relationship that we are trying to demonstrate. However, it is interesting to note that the points in the duration over time line plot where there are the biggest increases and decreases in duration do coorespond in a realistic way to the release of new technologies. For example, from the late 1960s to the early 1970s there was an increase in song duration, which most notably happens to be the year that the 8-track tape and the Turntable were released. Both of these technologies allowed people to play songs all the way through, or listen to records in their entirety - and sales in the music industry were dependent more on entire record sales, rather than specific songs. Therefore, it didn't matter if a song was 15 minutes or 30 seconds, what mattered more was that the record was good. A similar but opposite trend is shown from 2000 to the late 2000's, which is when the invention of new streaming services like Napstr, Itunes, Myspace were brought to market as well as the invention of new music playing inventions like the IPod. Following these inventions, the revenue model where artists were getting paid per stream of songs rather that records was gaining steam and people's attention spans were getting shorter. Therefore, it makes sense why song duration decreased because artists were spending less money making shorter songs (since they were getting paid by if the song was listened to regardless of length) and the consumer had a shorter attention span. Although year as a standalone variable is not a good predictor of song duration, it was overall interesting to see the coorelation between the rise of next technologies and the direct effect that they had on the duration of songs that were released. A more specific coorelation could be the amount of money artists make per song vs song duration over time, this would be a more accurate predictor that we can run linear regression on - and ideally, also be able to compare the means of random subsets of each dataframe. 

However, the reason that year might not be a good variable to predict duration of songs is because there are other outside factors that have influenced duration of music over time: such as streaming. In a real world context, this decrease can be attributed to the rise of the streaming service economic model. Artists get paid between 0.004 and 0.008 cents for every stream that their song recieves on a streaming platform. This model, compared to the model demonstrated with CD's and vinyl, disincentivizes artists to make longer individual songs. For example, in the age of CD and Vinyl an artist was incentivized to make the best songs in order to promote the sales of their physical records, they didn't make money off of individual songs. There are external theories claiming that as people's attention spans lower, artists are making songs that are more danceable and shorter, in order to appeal to the modern consumer. Through our data we discovered that danceability has increased over time as duration of songs has decreased - which gives us a fair amount of confidence that this theory could prove to be true.
